```python
import pickle
# LOAD PICKLE FILE
file_open = open("mnist_small.pkl", 'rb')
data = pickle.load(file_open)
print(data)
```

```python
# CREATE SEPARTE LIST FOR X (DATA POINT) AND Y (LABEL)
X = data['X']
Y = data['Y']
```

```python
# APPLY PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
transformed_data = pca.transform(X)
```

```python
import matplotlib.pyplot as plt
# PLOT SCATTER FOR PCA
colors = {0: 'red', 1: 'green', 2: 'yellow', 3: 'black', 4: 'brown', 5: 'pink', 6: 'purple', 7: 'magenta', 8: 'orange', 9: 'blue'}
for j in range(10):
  similar_x=[]
  similar_y=[]
  for i in range(len(Y)):
    if int(Y[i][0])==j:
      similar_x.append(transformed_data[i][0])
      similar_y.append(transformed_data[i][1])
  plt.scatter(similar_x,similar_y,c=colors[j])
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Plot with Colored Labels")
plt.legend()
plt.show()
```

```python
from sklearn.manifold import TSNE
# APPLY tSNE
tsne = TSNE(n_components=2)
transformed_data = tsne.fit_transform(X)
```

```python
colors = {0: 'red', 1: 'green', 2: 'yellow', 3: 'black', 4: 'brown', 5: 'pink', 6: 'purple', 7: 'magenta', 8: 'orange', 9: 'blue'}
# PLOT SCATTER FOR tSNE
for j in range(10):
  similar_x=[]
  similar_y=[]
  for i in range(len(Y)):
    if int(Y[i][0])==j:
      similar_x.append(transformed_data[i][0])
      similar_y.append(transformed_data[i][1])
  plt.scatter(similar_x,similar_y,c=colors[j])
plt.xlabel("TSNE Component 1")
plt.ylabel("TSNE Component 2")
plt.title("TSNE Plot with Colored Labels")
plt.legend()
plt.show()
```
